# DNA and RNA MPI Timing & Speedup Plots

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import scipy
sns.set_context('talk')
sns.set_style('ticks')
import warnings
warnings.filterwarnings("ignore")
```

```python
#cores and timings
nodes = [1, 2, 4, 8, 16, 32, 64, 128]
dna = [1622.531984, 801.062617, 411.091498, 304.68746, 114.665329, 62.339482, 39.894728, 28.847915]
rna = [2341.018516, 1050.454137, 515.89984, 263.873469, 144.668194, 78.627554, 46.580353, 42.151208]

#speedup
d_speedup = [dna[0]/p for p in dna]
r_speedup = [rna[0]/p for p in rna]

pd.DataFrame({'Cores':nodes, 'DNA Speedup': d_speedup, 'RNA Speedup':r_speedup})
```

```python
#plot time vs cores
plt.plot(nodes, dna, '-o', c = '#de2d26', label='DNA Sample 1')
plt.plot(nodes, rna, '-o', c = '#3182bd', label='RNA Sample 1')
plt.title('MPI One Chromosome Elapsed Time vs. Nodes')
plt.xlabel('Cores')
plt.ylabel('Elapsed Time [seconds]')
plt.legend(loc = 'lower left')
plt.xscale('log')
plt.yscale('log')
plt.xticks([1, 2, 4, 8, 16, 32, 64, 128], ('1','2','4','8','16','36','64','128'))
sns.despine();
```

```python
#plot speedup

# add diagonal line
x = [0, 110]
y = [0, 110]
plt.plot(x, y, '--')

#speedup vs cores
plt.plot(nodes, d_speedup, '-o', c = '#de2d26', label='DNA Sample 1')
plt.plot(nodes, r_speedup, '-o', c = '#3182bd', label='RNA Sample 1')
plt.title('MPI One Chromosome Speedup vs. Nodes')
plt.xlabel('Cores')
plt.ylabel('Speedup')
plt.legend(loc = 'upper left')
plt.xscale('log')
plt.yscale('log')
plt.xticks([1, 2, 4, 8, 16, 32, 64, 128], ('1','2','4','8','16','36','64','128'))
sns.despine();
```
