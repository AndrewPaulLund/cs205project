```python
DNA.counts<-read.csv("/Users/divyammisra/cs205project/data/HG00096.DNA.chr1.counts")
hist(DNA.counts$X0, breaks=10000, xlim=c(0,1000))

RNA.counts<-read.csv("/Users/divyammisra/cs205project/data/HG00096.RNA.chr1.counts")
hist(log10(RNA.counts$X0+1), breaks=10000, xlim=c(0,1000))
hist(log10(RNA.counts$X0+1), breaks=100000, xlim=c(0,500))


pdf("plots.pdf")
png("plots_%d.png")
# Plot hist
hist(log10(DNA.counts$X0+1), xlab="log10(Number of data points + 1)", col="#de2d26", main="DNA Dataset")
hist(log10(RNA.counts$X0+1), xlab="log10(Number of data points + 1)", col="#3182bd", main="RNA Dataset")

# Plot input data
plot(seq(length(log10(DNA.counts$X0+1))), log10(DNA.counts$X0+1), 
     col="#de2d26", pch=20, xlab="Order of input data",
     ylab="log10(Number of data points + 1)", main="DNA Data Heterogenity")
plot(seq(length(log10(RNA.counts$X0+1))), log10(RNA.counts$X0+1), 
     col="#3182bd", pch=20, xlab="Order of input data", 
     ylab="log10(Number of data points + 1)", main="RNA Data Heterogenity")


# Barplot
barplot(log10(DNA.counts$X0+1), xlab="Order of input data", 
        ylab="Number of reads (log)", main="DNA Sample 1 Heterogeneity", col="#de2d26", border="#de2d26")
barplot(log10(RNA.counts$X0+1), xlab="Order of input data", 
        ylab="Number of reads (log)", main="RNA Sample 1 heterogeneity", col="#3182bd", border="#3182bd")


dev.off()


x<-seq(length(log10(DNA.counts$X0+1)))
y<-log10(DNA.counts$X0+1)
lo<-loess(y ~ x, )
plot(lo)
```
